```python
# do work I need to do to get fits to spence model data

from aerosandbox import numpy as np
from dynamics.AeroModelling.SpenceModel import *
from dynamics.AeroModelling.propulsorModels import *
import matplotlib.pyplot as plt
from scipy.interpolate import make_splrep
from scipy.optimize import curve_fit
```

```python
# problem setup
Cj = np.linspace(0.01,10,100)
E = 0.3

```

```python
# plot components of D as function of Cj

Dm = np.zeros((9,len(Cj)))
for idx,val in enumerate(Cj):
    Dm[:,idx] = DcoeffSolution(9,E,val).flatten()
    
```

```python
plt.plot(Cj,Dm[1,:],label='D1')
plt.plot(Cj,Dm[2,:],label='D2')

plt.show()
```

```python
from aerosandbox import InterpolatedModel

# fit spline to Dm1

Dm1Interp = InterpolatedModel(Cj, Dm[1,:].flatten())

# compare
plt.plot(Cj,Dm[1,:],label='data',linestyle='solid',color='red')
plt.plot(Cj,Dm1Interp(Cj),label='data',linestyle='dotted',color='red',marker='s')

```

```python
# fit spline to Dm1

spl2 = make_splrep(Cj, Dm[2,:].flatten(), s=0)

# compare
plt.plot(Cj,Dm[2,:],label='data',linestyle='solid',color='red')
plt.plot(Cj,spl2(Cj),label='data',linestyle='dotted',color='red')
```
